```python
def separate_contigs(filename, verbose=False):
    if '.gz' in filename:
        with gzip.open(filename, 'rb') as f:
            file_content = f.read()
            
    else:
        with open(filename, 'rb') as f:
            file_content = f.read()
    f.close()
    file_content = file_content.decode('utf-8')
    current_string = file_content
    done = False
    genomes = {}
    while not done:
        idx_start = current_string.find('\n')
        header = current_string[:idx_start]
        idx_end = current_string[1:].find('>')
        if(idx_end==-1):#no more headers
            done=True
        genome = current_string[idx_start:idx_end]
        genome = genome.replace("\n", "")
        print(genome[:10])
        genomes[header]=genome
        current_string = current_string[idx_end+2:]
        if verbose:
            print('remaining string length:', len(current_string))
            print(repr(header))
    return genomes
genomes = separate_contigs('SmitsSA_2017/SmitsSA_2017__TZ_65642.fa')
```

```python
import os
from tables import *
import gzip
def retrieve_genomes(path, gzipped=False):
    if(gzipped):
        with gzip.open(path, 'r') as f:
            file_content = f.readlines()
            file_content = [content.strip() for content in file_content]
            
    else:
        with open(path, 'r') as f:
            file_content = f.readlines()
            file_content = [content.strip() for content in file_content]
    if isinstance(file_content[0], bytes):
        file_content = [line.decode('utf') for line in file_content]
    genomes = {}
    bases = ""
    header = file_content[0]
    
    for i in range(1, len(file_content)):
        if(">") in file_content[i]:
            genomes[header] = bases
            header = file_content[i]
            bases=""
        else:
            bases+= file_content[i] 
            if(i==len(file_content)-1):
                genomes[header] = bases
                header = file_content[i]
                bases=""
    return genomes

def tokenize_from_file(sample_title, folder='megahit_raw_out', out='subword',path='/gpfs/data/johnsonslab/nlp-genomics/t2d'):
#     print("tokenizing", sample_title)
    file = os.path.join(path, str(sample_title), folder, 'final.contigs.fa')
    genomes = retrieve_genomes(file).values()
    tokens = []
    assert out in ['subword', 'id'], "out must be 'id' or 'subword'"
    if out == 'subword':
        output_type = yttm.OutputType.SUBWORD
    elif out=='id':
        output_type = yttm.OutputType.ID
    for genome in genomes:
        tokens += bpe.encode(genome, output_type =output_type)
    return tokens
def get_tokenization(sample, path_to_db):
    with open_file(path_to_db, 'r') as file:
        data = file.root.data
        tkns=data[f'Tknz{sample}'].read()
    return tkns
# def tokenize(sample_title, folder, out='subword',path='/gpfs/data/johnsonslab/nlp-genomics/alzheimers/ferreiro-2023'):
#     tokens = data['Tknz%d' %sample_title].read()
#     return tokens
# total_lens = []
# with open_file('/gpfs/data/johnsonslab/nlp-genomics/alzheimers/ferreiro-2023/raw_tokenizations_8192k.h5', 'r') as file:
#     data = file.root.data
#     for sample in tqdm(key_df['sample_title']):
#         try:
#             tkns = data['Tknz%d' % sample].read()
#             total_lens.append(len(tkns))
#         except:
#             print(sample)
# print(max(total_lens))
# print(np.mean(total_lens))
# print(np.median(total_lens))
```

```python
def logit_pvalue(model, x):
    """ Calculate z-scores for scikit-learn LogisticRegression.
    parameters:
        model: fitted sklearn.linear_model.LogisticRegression with intercept and large C
        x:     matrix on which the model was fit
    This function uses asymtptics for maximum likelihood estimates.
    """
    p = model['clf'].predict_proba(x)
    n = len(p)
    m = len(model['clf'].coef_[0]) + 1
    coefs = np.concatenate([model['clf'].intercept_, model['clf'].coef_[0]])
    # print(np.array(x).shape)
    x_full = np.matrix(np.insert(np.array(x.todense()), 0, 1, axis = 1))
    ans = np.zeros((m, m))
    print(n)
    for i in range(n):
        ans = ans + np.dot(np.transpose(x_full[i, :]), x_full[i, :]) * p[i,1] * p[i, 0]
    vcov = np.linalg.inv(np.matrix(ans))
    se = np.sqrt(np.diag(vcov))
    t =  coefs/se  
    p = (1 - norm.cdf(abs(t))) * 2
    return p
```

# Repeating pipeline with thresholded tokens

```python
from tables import *
def get_tokenization(sample, path_to_db):
    with open_file(path_to_db, 'r') as file:
        data = file.root.data
        tkns=data['Tknz%d' % sample].read()
    return tkns
```

```python
import os
import pandas as pd
from functools import partial

path = '/gpfs/data/johnsonslab/nlp-genomics/alzheimers/ferreiro-2023'
key_df = pd.read_csv(os.path.join(path, 'key_df.csv'))
##need to remove the last entry due to the operation taking a long time
# key_df = key_df.iloc[:-1]
#need to remove missing index too
# idx = key_df[key_df['sample_title']==61707].index[0]
# key_df = pd.concat([key_df.iloc[:idx], key_df.iloc[idx + 1:]])
# tokenize_raw = partial(tokenize, folder='megahit_raw_out')
tokenize_kneaded = partial(get_tokenization, path_to_db=os.path.join(path, 'kneaded_tokenizations_threshold_10_8192k.h5'))
key_df
```

## Preparing model

```python
%%time
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, accuracy_score
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.decomposition import TruncatedSVD

import time
import numpy as np
np.random.seed(42)

X_train, X_test, y_train, y_test = train_test_split(key_df['sample_title'], key_df['disease_state'])
# tf_idf_transformer = TfidfTransformer()
# start = time.time()
# sample = [tokenize_raw(x) for x in X_train[:2]]
# print(time.time()-start)
# tf_idf_transformer.fit(sample)
# X_train = key_df['sample_title'][:2]
tf_idf_vectorizer = TfidfVectorizer(tokenizer=tokenize_kneaded, lowercase=False)
svd_components = 100
# tf_idf_vectorizer_fit = tf_idf_vectorizer.fit(X_train)
classifier_rf = Pipeline([('tfidf',tf_idf_vectorizer),
                          ('svd', TruncatedSVD(n_components=svd_components, n_iter=7, random_state=42)),
('clf',RandomForestClassifier(bootstrap= False, criterion= 'entropy', n_estimators= 15, ))])
```

```python
def analyze(clf):
    #Predicting
    y_pred = clf.predict(X_test)
    yt_pred = clf.predict(X_train)
    #Analyzing
    cm = confusion_matrix(y_test,y_pred)
    print(f'Confusion Matrix :\n {cm}\n')
    print(f'Test Set Accuracy Score :\n {accuracy_score(y_test,y_pred)}\n')
    print(f'Train Set Accuracy Score :\n {accuracy_score(y_train,yt_pred)}\n')
    print(f'Classification Report :\n {classification_report(y_test,y_pred)}')
```

```python
%%time
start = time.time()
classifier_rf.fit(X_train, y_train)
print("fit time:", time.time()-start)
analysis_start = time.time()
analyze(classifier_rf)
print("analysis time:", time.time()-analysis_start)
# print("total time:", time.time()-start)
```

### Assigning lines in files to corpus chunks

```python
import numpy as np
from tqdm import tqdm
rng = np.random.RandomState(42)
for i in range(64):
    f = open(os.path.join(path, 'corpus_chunks', 'raw', f't2d_raw_chunk_{i}.txt'), 'wt')
    f.close()
with open(os.path.join(path, 'corpus_chunks', 'raw', 't2d_raw_all_seqs.txt'), 'wt') as g:
    for sample in tqdm(meta_df['Sample Id']):
        genomes =  retrieve_genomes(os.path.join(path, sample, 'megahit_raw_out', 'final.contigs.fa')).values()
        for genome in genomes:
            idx = rng.randint(0, 64) #get random file index
            f = open(os.path.join(path, 'corpus_chunks', 'raw', f't2d_raw_chunk_{idx}.txt'), 'at') #open random file
            f.write(genome + "\n") #write one line
            g.write(genome+"\n") #write to master corpus
            f.close()
```
